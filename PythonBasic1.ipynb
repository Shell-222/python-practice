```python
# 7. Write a Python program to accept a filename from the user and print the extension of that. Go to the editor
# Sample filename : c
# Output : java

# filename = input("Type a filename with extension")
filename = "abc.1.2.3.java"

# Any . in the file's name? 
# Maybe search . from the back of string

# Hard coded way first, for concept proof
idx = filename.find(".")  # find the first "."
print(filename[idx+1:])


# Dynamic Code
filename_rvs = filename[::-1]
idx_rvs = filename_rvs.find(".")
print(filename_rvs[:idx_rvs][::-1])
```

```python
# 10. Write a Python program that accepts an integer (n) and computes the value of n+nn+nnn. Go to the editor
# Sample value of n is 5
# Expected Result : 615

n = 5

# Convert int to string
N = str(n)

# Cat the string "nn" and "nnn"
NN = N + N
NNN = NN + N

# Convert "nn" and "nnn" to int
nn = int(NN)
nnn = int(NNN)

# Sum up and print
print(n+nn+nnn)
```

```python
# 14. Write a Python program to calculate number of days between two dates.
# Sample dates : (2014, 7, 2), (2014, 7, 11)
# Expected output : 9 days

from datetime import date
f_date = date(2014, 7, 2)
l_date = date(2014, 7, 11)
delta = l_date - f_date
print(delta.days)
```

```python
# 17. Write a Python program to test whether a number is within 100 of 1000 or 2000. Go to the editor

def near_thousand(n):
      return ((abs(1000 - n) <= 100) or (abs(2000 - n) <= 100))
print(near_thousand(1000))
print(near_thousand(900))
print(near_thousand(800))   
print(near_thousand(2200))
```
